```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
##Wikipedia titles
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
soup=BeautifulSoup(page.content)
soup
title=[]
for i in soup.find_all('span',class_="mw-headline"):
    title.append(i.text)
title
df=pd.DataFrame({'Title':title})
df
```

```python
#IMDB top rated 100 movies
IMDB_page=requests.get('https://www.imdb.com/list/ls091520106/')
IMDB_page
content=BeautifulSoup(IMDB_page.content)
content
top_100_movies=[]
year=[]
rating=[]
for i in content.find_all('h3',class_='lister-item-header'):
    top_100_movies.append(i.find("a").text.replace("\n"," "))
top_100_movies

year=[]
for j in content.find_all('span',class_='lister-item-year text-muted unbold'):
    year.append(j.text)
year
rating=[]
for k in content.find_all('div',class_='ipl-rating-star small'):
    rating.append(k.text.replace("\n",""))
rating
data = pd.DataFrame()
data['Movie Title']=top_100_movies
data['Year']=year
data['Rating']=rating
data
```

```python
##Top 100 Indian Movies- IMDB
Indian_movie=requests.get('https://www.imdb.com/list/ls056092300/?sort=user_rating,desc&st_dt=&mode=detail&page=1&ref_=ttls_ref_rt_vt')
Indian_movie
page_content=BeautifulSoup(Indian_movie.content)
page_content
movie_title=[]
for i in page_content.find_all('h3',class_='lister-item-header'):
    movie_title.append(i.find('a').text)
movie_title
Release_Year=[]
for j in page_content.find_all('span',class_='lister-item-year text-muted unbold'):
    Release_Year.append(j.text)
Release_Year
IMDB_Rating=[]
for k in page_content.find_all('div',class_='ipl-rating-star small'):
    IMDB_Rating.append(k.text.replace("\n",""))
IMDB_Rating
data=pd.DataFrame()
data['movie_title']=movie_title
data['Release_Year']=Release_Year
data['Rating']=IMDB_Rating
data
```

```python
##former Presidents of India
URL=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
URL
page_content=BeautifulSoup(URL.content)
page_content
former_Presidents=[]
for i in page_content.find_all('div',class_="presidentListing"):
    former_Presidents.append(i.find("h3").text.split("(")[0])
former_Presidents
Term=[]
for j in page_content.find_all("span",class_="terms"):
    Term.append(j.text)
Term
data=pd.DataFrame()
data['President']=former_Presidents
data['Term of Service']=pd.Series(Term)
data
```

```python
##CNBC News Report
CNBC_URL=requests.get("https://www.cnbc.com/world/?region=world")
CNBC_URL
content=BeautifulSoup(CNBC_URL.content)
content
Headline=[]
for i in content.find_all('a',class_="LatestNews-headline"):
    Headline.append(i.text)
Headline
Time=[]
for j in content.find_all('span',class_="LatestNews-wrapper"):
    Time.append(j.text)
Time
News_Link=[]
for l in content.find_all('a',class_="LatestNews-headline"):
    News_Link.append(l['href'])
News_Link
data=pd.DataFrame()
data['HeadLine']=Headline
data['Time']=pd.Series(Time)
data['New_Link']=News_Link
data
```

```python
##Most Downloaded articles
article_URL=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
article_URL
page_content1=BeautifulSoup(article_URL.content)
page_content1
Paper_title=[]
for i in page_content1.find_all('h2',class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR"):
    Paper_title.append(i.text)
Paper_title
Author=[]
for j in page_content1.find_all('span',class_="sc-1w3fpd7-0 pgLAT"):
    Author.append(j.text)
Author
date=[]
for k in page_content1.find_all('span',class_="sc-1thf9ly-2 bKddwo"):
    date.append(k.text)
date
link=[]
for l in page_content1.find_all('a',class_="sc-5smygv-0 nrDZj"):
    link.append(l['href'])
link
df=pd.DataFrame()
df['Paper_Title']=Paper_title
df['Author']=Author
df['Published_Date']=date
df['Link']=link
df
```

```python
##Dine out
dine_out=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
dine_out
rest_content=BeautifulSoup(dine_out.content)
rest_content
rest_name=[]
for i in rest_content.find_all('a',class_="restnt-name ellipsis"):
    rest_name.append(i.text)
rest_name
location=[]
for j in rest_content.find_all('div',class_="restnt-loc ellipsis"):
    location.append(j.text)
location
rest_rating=[]
for k in rest_content.find_all('div',class_="restnt-rating rating-4"):
    rest_rating.append(k.text)
rest_rating
image=[]
for l in rest_content.find_all('img',class_="no-img"):
    image.append(l['data-src'])
image
df=pd.DataFrame({'Restaurant Name':rest_name,'Location':location,'rating':rest_rating,'Image URL':image})
df
```

```python
##Top Publications
scholar_form=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
scholar_form
content1=BeautifulSoup(scholar_form.content)
content1
Rank=[]
for i in content1.find_all('td',class_="gsc_mvt_p"):
    Rank.append(i.text.replace(".",""))
Rank
publications=[]
for j in content1.find_all('td',class_="gsc_mvt_t"):
    publications.append(j.text)
publications
h5_index=[]
for k in content1.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(k.text)
h5_index
h5_median=[]
for l in content1.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(l.text)
h5_median
df=pd.DataFrame({'Rank':Rank,'Publications':publications,'h5_index':h5_index,'h5_median':h5_median})
df
```
